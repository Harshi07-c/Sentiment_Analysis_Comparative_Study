```python
import pandas as pd
```

```python
df = pd.read_csv("/content/output (1).csv")
```

```python
display(df)
```

#Positive, Negative, and Neutral Counts:

```python
# Calculate counts for each sentiment
sentiment_counts = df[['vader', 'afinn', 'TextBlob', 'BERT', 'T5']].apply(lambda x: x.value_counts()).transpose()

# Display counts
print(sentiment_counts)
```

#Visualizations:

##Bar Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set up the figure
plt.figure(figsize=(10, 6))

# Plot the bar chart
sns.barplot(x=sentiment_counts.index, y='positive', data=sentiment_counts, color='green', label='Positive')
sns.barplot(x=sentiment_counts.index, y='negative', data=sentiment_counts, color='red', label='Negative')
sns.barplot(x=sentiment_counts.index, y='neutral', data=sentiment_counts, color='blue', label='Neutral')

# Add labels and legend
plt.xlabel('Sentiment Analysis Models')
plt.ylabel('Count')
plt.title('Sentiment Distribution Across Models')
plt.legend()

# Show the plot
plt.show()
```
