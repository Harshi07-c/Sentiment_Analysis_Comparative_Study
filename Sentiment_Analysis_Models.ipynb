```python
import pandas as pd

df = pd.read_csv('/content/BA_AirlineReviews_final.csv')

# Display the DataFrame
display(df)
```

DROPPING UNWANTED COLUMNS

```python
selected_columns = ['ReviewBody']
df = df[selected_columns]
```

```python
display(df)
```

#APPLYING VADER

```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer

nltk.download('vader_lexicon')
```

```python
vader = SentimentIntensityAnalyzer()
scores = []
for tweet in df["ReviewBody"] :
  score = vader.polarity_scores(tweet)
  scores.append(score)

vader_analysis = []
for i in scores :
  if i["compound"] >= 0.05 :
    vader_analysis.append("positive")
  elif i["compound"] <= -0.05 :
    vader_analysis.append("negative")
  else :
    vader_analysis.append("neutral")

df["vader"] = vader_analysis
display(df)
```

```python
from afinn import Afinn
from nltk.sentiment import SentimentIntensityAnalyzer

# Assuming df is your DataFrame with a column named "Text"

# Initialize the AFINN sentiment analyzer
afinn = Afinn()

# Apply AFINN analysis and store results in the DataFrame
df['afinn'] = df['ReviewBody'].apply(lambda x: afinn.score(x))

# Function to get sentiment based on AFINN score
def get_afinn_sentiment(afinn_score):
    if afinn_score >= 0.05:
        return 'positive'
    elif afinn_score <= -0.05:
        return 'negative'
    else:
        return 'neutral'

# Apply AFINN sentiment and store results in the DataFrame
df['afinn'] = df['afinn'].apply(lambda x: get_afinn_sentiment(x))

# Display the DataFrame with AFINN sentiment information
display(df)
```

#Text Blob

```python
from textblob import TextBlob

df['TextBlob'] = df['ReviewBody'].apply(lambda x: TextBlob(str(x)).sentiment.polarity)

df['TextBlob'] = df['TextBlob'].apply(lambda x: 'positive' if x >= 0.05 else ('negative' if x <= -0.05 else 'neutral'))

display(df)
```

#BERT

```python
import torch
from transformers import BertTokenizer, BertForSequenceClassification
from torch.utils.data import DataLoader
from torch.nn.functional import softmax
from tqdm import tqdm

# Load the pre-trained BERT model for sequence classification
model_name = 'nlptown/bert-base-multilingual-uncased-sentiment'
tokenizer = BertTokenizer.from_pretrained(model_name)
model = BertForSequenceClassification.from_pretrained(model_name)


reviews = df['ReviewBody'].tolist()

# Tokenize and encode the reviews
inputs = tokenizer(reviews, return_tensors='pt', truncation=True, padding=True, max_length=512)

# Wrap the data in DataLoader
dataset = torch.utils.data.TensorDataset(inputs['input_ids'], inputs['attention_mask'])
dataloader = DataLoader(dataset)

# Make predictions
model.eval()
predictions = []

with torch.no_grad():
    for batch in tqdm(dataloader, desc='Predicting'):
        inputs, attention_mask = batch
        outputs = model(inputs, attention_mask=attention_mask)
        logits = outputs.logits
        probabilities = softmax(logits, dim=1)
        predictions.extend(probabilities[:, 1].tolist())

# Add sentiment scores to DataFrame
df['BERT'] = predictions

# Classify sentiment into categories
df['BERT'] = df['BERT'].apply(lambda x: 'positive' if x >= 0.05 else ('negative' if x <= -0.05 else 'neutral'))

# Display DataFrame
print(df)
```

#T5

```python
import torch
from transformers import T5ForConditionalGeneration, T5Tokenizer
from tqdm import tqdm

# Load the pre-trained T5 model for conditional generation
model_name = 't5-base'
tokenizer = T5Tokenizer.from_pretrained(model_name)
model = T5ForConditionalGeneration.from_pretrained(model_name)

# Function to perform sentiment analysis using T5
def t5_sentiment_analysis(text):
    prompt = f"Sentiment of the following review is: \"{text}\""
    input_ids = tokenizer.encode(prompt, return_tensors="pt", max_length=512, truncation=True)

    # Generate text based on the input
    output = model.generate(input_ids, max_length=50, num_beams=5, no_repeat_ngram_size=2, top_k=50, top_p=0.95, temperature=0.7)

    # Decode the generated text
    generated_text = tokenizer.decode(output[0], skip_special_tokens=True)

    # Interpret sentiment based on generated text
    sentiment = 'positive' if 'positive' in generated_text.lower() else ('negative' if 'negative' in generated_text.lower() else 'neutral')

    return sentiment

# Apply the function to each review in the DataFrame
df['T5_Sentiment'] = df['ReviewBody'].apply(t5_sentiment_analysis)

# Display the DataFrame
print(df)
```
